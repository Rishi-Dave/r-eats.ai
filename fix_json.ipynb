```python
import json
import re

def convert_to_specific_json(input_filepath="restaurant_dataset.txt", output_filepath="restaurants_specific.json"):
    """
    Converts the text-based restaurant data to a JSON array with specific categories:
    name, cuisine, ambiance, and features.
    """
    restaurants_data = []
    try:
        with open(input_filepath, 'r') as infile:
            all_data = infile.read()
            restaurant_blocks = [block.strip() for block in all_data.strip().split('\n\n') if block.strip()]

            for block in restaurant_blocks:
                restaurant_info = {}
                lines = block.strip().split('\n')
                for line in lines:
                    name_match = re.match(r"^\s*Restaurant Name:\s*(.*)$", line)
                    print(name_match)
                    if name_match:
                        restaurant_info['name'] = name_match.group(1).strip()
                        continue

                    cuisine_match = re.match(r"^\s*Cuisine:\s*(.*)$", line)
                    if cuisine_match:
                        restaurant_info['cuisine'] = [c.strip() for c in cuisine_match.group(1).split(',')]
                        continue

                    ambiance_match = re.match(r"^\s*Ambiance:\s*(.*)$", line)
                    if ambiance_match:
                        restaurant_info['ambiance'] = [a.strip() for a in ambiance_match.group(1).split(',')]
                        continue

                    features_match = re.match(r"^\s*Features:\s*(.*)$", line)
                    if features_match:
                        restaurant_info['features'] = [f.strip() for f in features_match.group(1).split(',')]
                        continue



                if 'name' in restaurant_info:
                    restaurants_data.append(restaurant_info)

        with open(output_filepath, 'w') as outfile:
            json.dump(restaurants_data, outfile, indent=4)
        print(restaurants_data)
        print(f"Successfully converted data to '{output_filepath}' with specific categories.")

    except FileNotFoundError:
        print(f"Error: Input file '{input_filepath}' not found.")
    except Exception as e:
        print(f"An error occurred during conversion: {e}")

# Example usage:
convert_to_specific_json()
```
